```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = '/content/drive/MyDrive/1.FACUL/2023_1/Lab controle e servo mecanismo/tabela_6.xlsx'
df = pd.read_excel(url)

# Substituir ',' por '.' nas colunas 'Vref' e 'VL'
df['Vtaco'] = df['Vtaco'].replace(',', '.')
df['ω'] = df['ω'].replace(',', '.')

# Converter as colunas para tipo numérico
df['Vtaco'] = pd.to_numeric(df['Vtaco'])
df['ω'] = pd.to_numeric(df['ω'])

# Plotar o gráfico
plt.scatter(df['ω'], df['Vtaco'])
plt.ylim(0, 6)  # Definir escala do eixo y
plt.xlim(0, 0.4)   # Definir escala do eixo x
plt.xlabel('ω')
plt.ylabel('Vtaco')

# Traçar uma reta que se aproxima ao máximo de todos os pontos
coefficients = np.polyfit(df['ω'], df['Vtaco'], 1)
polynomial = np.poly1d(coefficients)
x = np.linspace(0, 1, 100)
plt.plot(x, polynomial(x), color='red')

# Calcular o coeficiente angular da reta
coef_angular = coefficients[0]
plt.text(0.01, 5.5, f'Coeficiente Angular: {coef_angular:.2f}', fontsize=12)

# Exibir o gráfico
plt.show()
```
